```python
# Shabang

import matplotlib.pyplot as plt 
import pandas as pd
import streamlit as st
import numpy as np
import matplotlib
import seaborn as sns 
```

```python
st.set_option('deprecation.showPyplotGlobalUse', False)
st.title("World Bank Reserves")
```

```python
df_data=pd.read_csv('COFER_12-27-2020 06-41-57-24.csv')

df_data.describe()
```

```python
df_data.columns
```

```python
df_data.head(3)
```

```python
df_data['Unnamed: 7'].unique()
```

```python
df_data=df_data.drop(columns=['Status','Unnamed: 7'],axis=1)
```

```python
df_data.head()
```

```python
df_data[['Reserve_Type','Detail_Reserve','Reserve_Unit']]=df_data['Indicator Name'].str.split(",",expand=True)
df_data=df_data.drop(columns=['Indicator Name'],axis=1)
```

```python
df_data['Country Name'].unique()
```

```python
df_data['Reserve_Type'].unique()
```

```python
World_Economy=df_data[df_data['Country Name'] == 'World']
```

```python
World_Economy['Country Code'].unique()
```

```python
World_Economy=World_Economy.drop(columns=['Country Code'],axis=1)
World_Economy.columns
```

```python
World_Economy[['Time Period','Value','Reserve_Type','Detail_Reserve']].sort_values(['Time Period'],ascending=[True])
```

```python
currency_reserve=World_Economy[World_Economy.Reserve_Type.isin(['Allocated Reserves'])
                               & World_Economy.Detail_Reserve.str.contains('US Dollars')]
currency_reserve
```

```python
currency_reserve=currency_reserve[['Time Period','Value','Detail_Reserve']].sort_values(
    ['Time Period'],ascending=[True])
currency_reserve['Value']=(currency_reserve['Value']/(10**12))
currency_reserve=currency_reserve.set_index('Time Period')
st.table(currency_reserve)
currency_reserve
```

```python
st.header("World Bank Reserve in USD")
```

```python
st.subheader("Bar Plot")
currency_reserve.plot(kind='bar')
st.pyplot()
```

```python
#joinplot
st.subheader("JointPlot")
sns.jointplot(x='Time Period',y='Value',data=currency_reserve,kind='scatter')
st.pyplot()
```

```python
#Rugplot
st.subheader("Rugplot")
sns.rugplot(x='Time Period',y='Value',data=currency_reserve)
st.pyplot()
```
